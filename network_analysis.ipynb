```python
import pandas as pd
import numpy as np
import operator
import networkx as nx
```

```python
infile = 'User_Edge_df.csv'
df_edges = pd.read_csv(infile)

infile = 'USER_ID_df.csv'
df_nodes = pd.read_csv(infile, index_col=0)
```

```python
df_edges
```

```python
# Remove original tweets
df_edges = df_edges[df_edges.Target!=0]
```

```python
G = nx.from_edgelist(df_edges[['Source','Target']].values)
```

```python
(12301, 25383)
```

```python
G.number_of_nodes(), G.number_of_edges()
```

```python
# The bridging centrality of a node is defined as its betweenness centrality times its bridging coefficient,
# and the bridging coefficient is the sum of the degrees of a node's neighbours divided by that node's degree.

def bridging_coefficient(G, n):
    top    = sum( G.degree(i) for i in G.neighbors(n) )
    bottom = G.degree(n)
    return top/bottom

bridging_coefs = { n:bridging_coefficient(G, n) for n in G.nodes() }
```

```python
betweenness_centralities = nx.betweenness_centrality(G)
```

```python
bridging_centrality = { n:(bridging_coefs[n]*betweenness_centralities[n]) for n in G.nodes() }
```

```python
eigenvector_centrality = nx.eigenvector_centrality(G)
degree_centrality = nx.degree_centrality(G)
```

```python
node_with_max_betweenness_centrality = max(betweenness_centralities.items(), key=operator.itemgetter(1))[0]
node_with_max_bridging_centrality    = max(bridging_centrality.items(), key=operator.itemgetter(1))[0]
node_with_max_eigenvector_centrality = max(eigenvector_centrality.items(), key=operator.itemgetter(1))[0]
node_with_max_degree_centrality      = max(degree_centrality.items(), key=operator.itemgetter(1))[0]

node_with_max_betweenness_centrality, node_with_max_bridging_centrality, node_with_max_eigenvector_centrality, node_with_max_degree_centrality
```

```python
print(node_with_max_betweenness_centrality, df_nodes.loc[node_with_max_betweenness_centrality])
print(node_with_max_bridging_centrality, df_nodes.loc[node_with_max_bridging_centrality])
print(node_with_max_eigenvector_centrality, df_nodes.loc[node_with_max_eigenvector_centrality])
print(node_with_max_degree_centrality, df_nodes.loc[node_with_max_degree_centrality])
```
